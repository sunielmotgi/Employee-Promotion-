```python
import numpy as np
import pandas as pd
```

```python
df2= pd.read_csv('cleandempsm_data.csv')
df2
```

```python
df3=df2[(df2['is_promoted'] > 0)]
df3
```

```python
df2
```

```python
df3['recruitment_channel'].unique()
```

```python
df3.filter(items=['is_promoted'])
```

```python
X = df3.iloc[:,1:12]
y = df3['is_promoted']
```

```python
X.head(1)
```

```python
y
```

```python
from sklearn_pandas import DataFrameMapper
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
dm = DataFrameMapper([(['department',LabelEncoder()]),(["education", LabelEncoder()]), (["gender",LabelEncoder()])])
```

```python
dm
```

```python
ml = []
ml.append(['mapper',dm])
ml.append(['SC',StandardScaler()])
ml.append(['RF',RandomForestClassifier()])
```

```python
from sklearn.pipeline import Pipeline
import warnings
warnings.filterwarnings('ignore')
```

```python
model = Pipeline(ml)
model.fit(X,y)
```

```python
dict(df3.iloc[10,1:])
```

```python
new ={'department': 'Technology',
 'education': "Master's & above",
 'gender': 'm',
 'no_of_trainings': 1,
 'age': 34,
 'previous_year_rating': 5.0,
 'length_of_service': 7,
 'awards_won': 0,
 'avg_training_score': 78.0
  }
```

```python
aa = pd.DataFrame(new,index=[0])
aa
```

```python
prediction = model.predict(aa)
print(prediction)
if (prediction [0] == 0):
    print("Is not promoted")
else:
    print("Is promoted")
```
